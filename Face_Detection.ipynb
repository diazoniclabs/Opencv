```python
api_key = ''
api_secret = ''
image_url = 'https://st4.depositphotos.com/12982378/22072/i/600/depositphotos_220729084-stock-photo-smiling-adult-man-crossed-arms.jpg'
```

```python
import requests
request_url = "https://api-us.faceplusplus.com/facepp/v3/detect"
url = f"{request_url}?api_key={api_key}&api_secret={api_secret}&image_url={image_url}"


response = requests.request("POST",url)
d = response.json()['faces'][0]['face_rectangle']
```

```python
img = cv2.imread('person.jpg')
x = d['left']
y = d['top']
h = d['height']
w = d['width']
crop_face = img[y:y+h,x:x+w]
cv2.imwrite(f'Output.jpg',crop_face)
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,255),3)
cv2.putText(img,"Person",(x,y),cv2.FONT_HERSHEY_DUPLEX,1,(255,0,0))
cv2_imshow(img)
```
